```python
X = data[['Daily Cases', 'Daily Deaths']]
y_cases = data['Daily Cases']
y_deaths = data['Daily Deaths']
# Create lag features
for lag in range(1, 8):  # 7 lag features (previous 7 days)
    data[f'Cases_Lag_{lag}'] = data['Daily Cases'].shift(lag)
    data[f'Deaths_Lag_{lag}'] = data['Daily Deaths'].shift(lag)

# Drop rows with NaN values created due to lagging
data.dropna(inplace=True)

# Feature variables (lag features)
X_cases = data[[f'Cases_Lag_{i}' for i in range(1, 8)]]
X_deaths = data[[f'Deaths_Lag_{i}' for i in range(1, 8)]]

# Target variables
y_cases = data['Daily Cases']
y_deaths = data['Daily Deaths']
# Split into train and test sets (80% train, 20% test)
X_train_cases, X_test_cases, y_train_cases, y_test_cases = train_test_split(X_cases, y_cases, test_size=0.2, shuffle=False)
X_train_deaths, X_test_deaths, y_train_deaths, y_test_deaths = train_test_split(X_deaths, y_deaths, test_size=0.2, shuffle=False)
# Linear Regression Model for Daily Cases
lr_model_cases = LinearRegression()
lr_model_cases.fit(X_train_cases, y_train_cases)

# Linear Regression Model for Daily Deaths
lr_model_deaths = LinearRegression()
lr_model_deaths.fit(X_train_deaths, y_train_deaths)
# Random Forest Regressor for Daily Cases
rf_model_cases = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_cases.fit(X_train_cases, y_train_cases)

# Random Forest Regressor for Daily Deaths
rf_model_deaths = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_deaths.fit(X_train_deaths, y_train_deaths)

# Predictions for Daily Cases
lr_predictions_cases = lr_model_cases.predict(X_test_cases)
mse_lr_cases = mean_squared_error(y_test_cases, lr_predictions_cases)
r2_lr_cases = r2_score(y_test_cases, lr_predictions_cases)

# Predictions for Daily Deaths
lr_predictions_deaths = lr_model_deaths.predict(X_test_deaths)
mse_lr_deaths = mean_squared_error(y_test_deaths, lr_predictions_deaths)
r2_lr_deaths = r2_score(y_test_deaths, lr_predictions_deaths)

print(f"Linear Regression MSE (Cases): {mse_lr_cases}")
print(f"Linear Regression R (Cases): {r2_lr_cases}")
print(f"Linear Regression MSE (Deaths): {mse_lr_deaths}")
print(f"Linear Regression R (Deaths): {r2_lr_deaths}")
# Predictions for Daily Cases
rf_predictions_cases = rf_model_cases.predict(X_test_cases)
mse_rf_cases = mean_squared_error(y_test_cases, rf_predictions_cases)
r2_rf_cases = r2_score(y_test_cases, rf_predictions_cases)

# Predictions for Daily Deaths
rf_predictions_deaths = rf_model_deaths.predict(X_test_deaths)
mse_rf_deaths = mean_squared_error(y_test_deaths, rf_predictions_deaths)
r2_rf_deaths = r2_score(y_test_deaths, rf_predictions_deaths)

print(f"Random Forest MSE (Cases): {mse_rf_cases}")
print(f"Random Forest R (Cases): {r2_rf_cases}")
print(f"Random Forest MSE (Deaths): {mse_rf_deaths}")
print(f"Random Forest R (Deaths): {r2_rf_deaths}")
```

```python
# Ensure the index is in datetime format
data.index = pd.to_datetime(data.index)

# Generate future dates starting from the last date + 1 day
last_date = data.index[-1]  # Ensure this is datetime
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=7, freq='D')

# Initialize lists to hold predictions
future_cases = []
future_deaths = []

# Use the last 7 days of cases and deaths as input
last_7_days_cases = data[[f'Cases_Lag_{i}' for i in range(1, 8)]].tail(1).copy()
last_7_days_deaths = data[[f'Deaths_Lag_{i}' for i in range(1, 8)]].tail(1).copy()

# Iteratively predict the next 7 days
for i in range(7):
    # Predict the next day's cases and deaths
    predicted_cases = rf_model_cases.predict(last_7_days_cases)[0]
    predicted_deaths = rf_model_deaths.predict(last_7_days_deaths)[0]
    
    # Store the predictions
    future_cases.append(predicted_cases)
    future_deaths.append(predicted_deaths)
    
    # Update the lagged features by shifting the input and adding the predicted value
    last_7_days_cases = last_7_days_cases.shift(-1, axis=1)
    last_7_days_cases.iloc[0, -1] = predicted_cases  # Add the new prediction
    
    last_7_days_deaths = last_7_days_deaths.shift(-1, axis=1)
    last_7_days_deaths.iloc[0, -1] = predicted_deaths  # Add the new prediction

# Output the predictions
for i, date in enumerate(future_dates):
    print(f"Predicted cases for {date.date()}: {future_cases[i]}")
    print(f"Predicted deaths for {date.date()}: {future_deaths[i]}")
```
